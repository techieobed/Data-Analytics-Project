Step 1: Business Understanding

Introducation

A "startup" is often defined by its age, size, and funding mechanism, although there is no exact definition. A startup is a fledgling company that is only a few years old and has yet to generate continuous revenue. These startups operate on a modest scale, often with only a working prototype or a paid pilot, but they have the potential for quick development and expansion. Initially, they are sponsored by the founders' personal networks, including friends and family, and they aggressively seek more capital to support their growth and establish a viable business.


Objective

The primary goal is to analyze the funding ecosystem for start-ups in India from 2018 to 2021, with an emphasis on finding major trends and patterns. This report will provide critical insights into the Indian startup ecosystem, guiding strategic decisions for successfully entering and navigating the market.

Hypothesis:

a. Rationale:Examining the allocation of money across the various stages of start-ups helps reveal where the majority of investments are going. This information is useful for new start-ups as they analyze their possible funding opportunities at different phases of development.

b. Null Hypothesis: From 2018 to 2021, the proportion of total funding obtained by early-stage start-ups (seed and series A) did not differ considerably from that received by later-stage start-ups (series B and beyond).

c. Alternative Hypothesis 

From 2018 to 2021, early-stage start-ups (seed and series A) received much more funding than later-stage start-ups (series B and beyond).


Business Questions
1. Who were the most active investors in India's start-up scene?

Identifying the most active investors helps you understand who is driving the start-up ecosystem. This information is critical for creating strategic collaborations, comprehending investment trends, and finding potential financing sources.
i. Ranking: Place investors in order of their number of transactions.ii. Visualization: Use bar charts and heatmaps to display the most active investors and their investment patterns by year, total funding amount, and average funding amount.iii. Network Analysis: Conduct a network analysis to better understand the linkages and patterns of co-investment among top investors.


2. Is there a geographic pattern or hotspot for start-up investment in India?

Identifying geographic trends or hotspots for start-up funding aids in understanding the regional distribution of investments. This data is useful for determining investment targets, identifying regional innovation hotspots, and developing market entry strategies.

i. Visualization: Use heatmaps to show funding concentrations in various places. ii. Cluster Analysis: Use cluster analysis to determine which regions have a high concentration of start-up funding. iii. Comparative Analysis: Compare funding levels across regions to identify hotspots.

3. Is there an annual pattern in Indian start-up investment from 2018 to 2021?

Evaluating annual patterns in start-up funding allows you to evaluate the growth and dynamics of the start-up ecosystem over time. This data is essential for strategic planning and anticipating future trends, as it provides insights into how the investment climate has changed.
i. Trend Analysis: Use line charts to depict year-over-year changes in total funding amounts. ii. Statistical Analysis: Determine annual growth rates and conduct trend analysis to discover any noteworthy upward or downward patterns.

4. Which industries received the most funding each year?
Examining the industries with the most funding can provide extra context for assessing the influence of major investors.

5. How do a start-up's years of existence or age affect funding opportunities?
Understanding how a start-up's age affects investment prospects can shed light on whether start-ups financed by top investors have larger subsequent financing rounds.

6.What funding stages (seed, series A, etc.) garnered the most investment?
Examining the funding stages with the most investments can provide insight into prominent investors' funding practices.
